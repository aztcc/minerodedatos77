```python
import pandas as pd

df = pd.read_csv('items.csv')

df.eval('importe = precio * cantidad', inplace=True)

dfa = df.query("categoria == 'A'")

promedio_importe_por_categoria = df.groupby('categoria')['importe'].mean()

promedio_a = promedio_importe_por_categoria.loc['A']
respuesta_numerica = round(promedio_a, 3)

print(f"El promedio de importe de la categoría A es: {respuesta_numerica}")
```

```python
import pandas as pd

dfc = pd.read_csv('clientes.csv')
dfo = pd.read_csv('ordenes.csv')

merged_df = pd.merge(dfc, dfo, on='cliente_id', how='left')

total_rows_merged = len(merged_df)
print(f"Número total de filas del left merge: {total_rows_merged}")
monto_total_por_cliente = merged_df.groupby('cliente_id')['monto'].sum().fillna(0)

monto_cliente_1 = monto_total_por_cliente.loc[1]
print(f"El monto total por cliente_id=1 es: {monto_cliente_1}")
```

```python
import pandas as pd

df = pd.read_csv('sensores.csv')
df = df[['ts', 'temperatura', 'humedad']]

df['ts'] = pd.to_datetime(df['ts'], format='mixed', errors='coerce')

df['temperatura'] = pd.to_numeric(df['temperatura'], errors='coerce')
df['humedad'] = pd.to_numeric(df['humedad'], errors='coerce')

mediana_temperatura = df['temperatura'].median()
media_humedad = df['humedad'].mean()

df['temperatura'].fillna(mediana_temperatura, inplace=True)
df['humedad'].fillna(media_humedad, inplace=True)

nat_count_ts = df['ts'].isnull().sum()
print(f"(a) Número de NaT en 'ts': {nat_count_ts}")

final_mediana_temperatura = round(df['temperatura'].median(), 1)
print(f"(b) Mediana final de temperatura: {final_mediana_temperatura}")

final_media_humedad = round(df['humedad'].mean(), 3)
print(f"(c) Media final de humedad: {final_media_humedad}")
```
