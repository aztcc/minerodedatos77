###Introducción

La minería de datos (o Data Mining) es el proceso técnico de explorar y analizar grandes volúmenes de datos para descubrir patrones, tendencias o reglas que no son evidentes a simple vista.

Este concepto puede ser confundido con la Big Data, pero la diferencia es que Big Data es el proceso de almacenar y manejar volumenes masivos de información, mientras que la mineria de datos es la herramienta que analiza estos datos para obtener la información que se busca.

####Ciencia de datos

La Ciencia de Datos es un campo multidisciplinario que usa métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento e ideas valiosas de grandes volúmenes de datos (Big Data), combinando estadística, matemáticas, informática e inteligencia artificial, con el fin de entender fenómenos, predecir tendencias y apoyar la toma de decisiones informadas en diversas áreas, desde negocios hasta ciencia.

####The Basics of NumPy Arrays

```python
import numpy as np
np.random.seed(0)  # seed for reproducibility

x1 = np.random.randint(10, size=6)  # One-dimensional array
x2 = np.random.randint(10, size=(3, 4))  # Two-dimensional array
x3 = np.random.randint(10, size=(3, 4, 5))  # Three-dimensional array
```

```python
print("x3 ndim: ", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size)
```

```python
print("dtype:", x3.dtype)
```

```python
print("itemsize:", x3.itemsize, "bytes")
print("nbytes:", x3.nbytes, "bytes")
```

####Promedio (average) y promedio ponderado

```python
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)
```
