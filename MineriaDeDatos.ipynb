#Mineria de Datos

##Primer Parcial

###Introducción

La minería de datos (o Data Mining) es el proceso técnico de explorar y analizar grandes volúmenes de datos para descubrir patrones, tendencias o reglas que no son evidentes a simple vista.

Este concepto puede ser confundido con la Big Data, pero la diferencia es que Big Data es el proceso de almacenar y manejar volumenes masivos de información, mientras que la mineria de datos es la herramienta que analiza estos datos para obtener la información que se busca.

La Ciencia de Datos es un campo multidisciplinario que usa métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento e ideas valiosas de grandes volúmenes de datos (Big Data), combinando estadística, matemáticas, informática e inteligencia artificial, con el fin de entender fenómenos, predecir tendencias y apoyar la toma de decisiones informadas en diversas áreas, desde negocios hasta ciencia.

```python
# =========================
# 1) Histograma de edades
# =========================
fig, ax = plt.subplots(figsize=(8,5))
bins = range(int(df['edad'].min())-1, int(df['edad'].max())+2)
ax.hist(df['edad'], bins=bins, color=accent, edgecolor='white', alpha=0.85)
mean_age = df['edad'].mean()
med_age = df['edad'].median()
ax.axvline(mean_age, color=accent2, linestyle='--', linewidth=2, label=f"Media: {mean_age:.2f}")
ax.axvline(med_age, color=accent3, linestyle=':', linewidth=2.5, label=f"Mediana: {med_age:.2f}")
ax.set_title('Distribución de edades')
ax.set_xlabel('Edad (años)')
ax.set_ylabel('Frecuencia')
ax.legend()
plt.tight_layout()
plt.show()
fig.savefig('01_hist_edades.png', dpi=180)
```

```python
# =======================
# 2) Boxplot de edades
# =======================
fig, ax = plt.subplots(figsize=(6,4))
ax.boxplot(df['edad'], vert=False, patch_artist=True,
           boxprops=dict(facecolor=accent, color=accent),
           medianprops=dict(color='white', linewidth=2),
           whiskerprops=dict(color=accent), capprops=dict(color=accent))
ax.set_title('Caja y bigotes: edad')
ax.set_xlabel('Edad (años)')
plt.tight_layout()
plt.show()
fig.savefig('02_box_edades.png', dpi=180)
```

```python
# ====================================================
# 3) Barras: frecuencias de Sí/No para Python/Colab/GitHub
# ====================================================
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Programó en Python','Usó Google Colab','Cuenta de GitHub'],
    [accent, accent2, accent3]
):
    counts = df[col].value_counts().reindex([True, False]).fillna(0).astype(int)
    labels = ['Sí','No']
    ax.bar(labels, counts, color=[color, neutral], edgecolor='white')
    ax.set_title(title)
    ax.set_ylim(0, max(counts.max(), 1)+2)
    for i, v in enumerate(counts):
        ax.text(i, v+0.1, str(v), ha='center', va='bottom', fontsize=9)
axes[0].set_ylabel('Número de estudiantes')
plt.tight_layout()
plt.show()
fig.savefig('03_barras_frecuencias.png', dpi=180)
```

```python
# =======================================
# 4) Mapas de calor 2x2 para combinaciones
# =======================================
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)
    t[0,1] = np.count_nonzero(~a_bool & b_bool)
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool)

    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')
    ax.set_title(f'Tabla 2×2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','Sí'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','Sí'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
anios_escolaridad = np.array([0, 3, 6, 9, 12, 16, 18])
ingreso_mensual_estimado_mx = np.array([12621, 15723, 19587, 24401, 30398, 40747, 47176])
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.
Utilizar el primer punto y el útlimo para trazar una recta:   
$$  m = \frac{y_{2} - y_{1}}{x_{1} - x_{2}} $$
$$  b = y_{1} -m * x_{1} $$

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1
print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
import matplotlib.pyplot as plt
# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos $(y_{pred})$: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error

$$ MSE = \frac{\Sigma_{i = 1}^{n}(y_{i} - \hat y_{i})^2}{n} $$


En donde
$MSE = $ error cuadratico medio

$m = $ número de datos.

$Y_{i} = $ valores observados.

$\hat Y_{i} = $ valores predichos.

```python

# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

Además del MSE (Error Cuadrático Medio), existen varios indicadores clave para evaluar modelos de regresión lineal, cada uno con un enfoque distinto sobre el rendimiento del modelo:    



*   **RMSE** (Raíz del Error Cuadrático Medio): Es la raíz cuadrada del MSE. A diferencia del MSE, el RMSE expresa el error en las mismas unidades que la variable objetivo, lo que facilita su interpretación. Es útil para penalizar errores grandes, similar al MSE.
*   **MAE** (Error Absoluto Medio): Calcula el promedio de las diferencias absolutas entre las predicciones y los valores reales. Es más robusto ante valores atípicos (outliers) que el MSE, ya que no eleva al cuadrado los errores grandes.
*   **$R^{2}$** (Coeficiente de Determinación): Indica qué proporción de la variación de la variable dependiente es explicada por el modelo. Va de 0 a 1; un valor más cercano a 1 indica un mejor ajuste.    
* **$R^{2}$** Ajustado: Similar al **$R^{2}$**, pero ajusta el valor según el número de predictores (variables independientes) en el modelo. Es más preciso que el **$R^{2}$** cuando se añaden variables innecesarias, ya que penaliza la complejidad excesiva.   
*   **MAPE** (Error Porcentual Absoluto Medio): Mide el error en términos porcentuales, lo que facilita entender la precisión relativa del modelo (ej. "el modelo se equivoca en un 5% en promedio").

¿Qué variables son nominales, ordinales, de intervalo y de razón? Da justificación breve.    
Las variables nominales son genero, ciudad e id_cliente, ya que representan categorías sin un orden inherente; en el caso de id_cliente funciona únicamente como identificador y no debe analizarse estadísticamente. Las variables ordinales son segmento (Estándar < Oro < Platino) y satisfaccion (Likert 1–5), porque poseen un orden natural entre categorías, aunque las distancias entre ellas no necesariamente son iguales. La variable de intervalo es temperatura_sala, ya que las diferencias entre valores son significativas, pero el cero es arbitrario (0 °C no implica ausencia de temperatura). Finalmente, las variables de razón son edad, ingreso_mensual, visitas_ultimo_anio y dias_desde_ultima_visita, porque cuentan con un cero absoluto y permiten interpretar proporciones (por ejemplo, el doble de ingreso o el doble de visitas).   

¿Qué medidas resumen usaste para cada tipo y por qué?    
Para variables nominales se utilizaron frecuencias y proporciones, ya que no es válido calcular medias o medianas en categorías sin orden. En variables ordinales se analizaron distribuciones de frecuencia y la mediana como medida de tendencia central, debido a que respeta el orden sin asumir intervalos iguales entre categorías. En la variable de intervalo (temperatura_sala) se emplearon media, desviación estándar, mínimo y máximo, porque las diferencias entre valores son interpretables aunque no las proporciones. En las variables de razón se calcularon media, mediana, desviación estándar, cuartiles e incluso el coeficiente de variación, ya que permiten análisis más completos incluyendo dispersión y comparaciones proporcionales.

Para tratarla, se dio mayor peso interpretativo a la mediana y al rango intercuartílico (IQR) en lugar de depender únicamente de la media. En un análisis más avanzado, podría aplicarse una transformación logarítmica para reducir la asimetría y mejorar la estabilidad de modelos estadísticos.

Asimismo, monitorear continuamente la satisfacción por segmento permitiría detectar posibles áreas de mejora en el servicio. Finalmente, debido a la asimetría en los ingresos, las decisiones estratégicas deberían basarse en métricas robustas como la mediana y no únicamente en el promedio, para evitar sesgos provocados por valores extremos.

###Correlación

¿La relación es lineal?   
Para determinar si la relación entre dos variables es lineal, se debe analizar principalmente un gráfico de dispersión (scatter plot). Si los puntos forman una tendencia aproximada en línea recta, ya sea ascendente o descendente, se puede considerar que la relación es lineal y es apropiado utilizar la correlación de Pearson. Sin embargo, si la relación presenta una forma curva (por ejemplo, en U o exponencial), la correlación de Pearson podría no detectarla adecuadamente, por lo que sería más conveniente utilizar la correlación de Spearman, que permite capturar relaciones monótonas no lineales.   
   
¿Hay outliers que la distorsionan?   
Es importante revisar si existen valores atípicos mediante diagramas de caja (boxplots) o inspeccionando el gráfico de dispersión. Los outliers pueden alterar significativamente el valor de la correlación de Pearson, aumentando o disminuyendo artificialmente su magnitud. Si se detectan valores extremos que influyen demasiado en el resultado, se recomienda considerar métodos más robustos como Spearman o Kendall, o bien aplicar transformaciones a los datos. Si al eliminar un solo punto la correlación cambia de forma considerable, el análisis no es robusto y debe revisarse con mayor cuidado.   

¿Existe un confusor?   
Se debe analizar si existe una tercera variable que esté influyendo simultáneamente en las dos variables estudiadas, generando una correlación aparente. Este fenómeno ocurre cuando ambas variables están relacionadas con un mismo factor externo, lo que puede producir una correlación espuria. Para identificar este problema se puede recurrir a la correlación parcial, controlando el efecto de la posible variable confusora. Además, es fundamental considerar el contexto del problema y el conocimiento del dominio para interpretar correctamente los resultados.   

¿Es razonable desde el dominio del problema?   
La interpretación de una correlación debe apoyarse en el conocimiento teórico o práctico del área de estudio. No toda correlación tiene sentido lógico o causal, por lo que es necesario evaluar si la relación encontrada es coherente con la realidad del fenómeno analizado. Una correlación puede ser estadísticamente significativa pero carecer de fundamento conceptual, lo que indicaría que podría tratarse de una coincidencia o de una relación espuria. Por ello, la estadística debe complementarse siempre con el análisis contextual.   

¿Tiene implicación para el modelo?   
La correlación tiene implicaciones directas en la construcción de modelos de minería de datos y machine learning, especialmente en modelos lineales. Si dos variables independientes presentan alta correlación entre sí, puede existir multicolinealidad, lo que genera inestabilidad en los coeficientes, dificultades de interpretación y aumento en los errores estándar. En estos casos, es recomendable analizar la matriz de correlación y calcular el Factor de Inflación de la Varianza (VIF), eliminando variables redundantes o aplicando técnicas como reducción de dimensionalidad o regularización para mejorar la estabilidad del modelo.

```python
import math

# Datos de entrenamiento
X = [(2,4), (4,4), (4,2), (6,2)]
y = ['Rojo', 'Rojo', 'Azul', 'Azul']

# Punto a clasificar
P = (3, 3)
```

```python
def distancia_euclidiana(p1, p2):
    return math.sqrt((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)
```

```python
distancias = []
for i in range(len(X)):
  d = distancia_euclidiana(P, X[i])
  distancias.append((d, y[i]))

# Ordenar por distancia
distancias.sort(key=lambda x: x[0])

print(distancias)
print("Clase asignada:", distancias[0][1])
```

###Ejercicio 2: k‑NN manual con k = 3

```python
from collections import Counter

k = 3
vecinos = distancias[:k]

clases = [clase for _, clase in vecinos]
resultado = Counter(clases).most_common(1)

print("Vecinos:", vecinos)
print("Clase asignada:", resultado[0][0])
```
