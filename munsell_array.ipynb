```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_rows = 6   # Show only 6 rows of output
pd.options.mode.chained_assignment = None  # Suppress SettingWithCopyWarning
munsell = pd.read_csv('munsell-array-fixed-choice.csv')
munsell
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
english_color_percentages = munsell['english_color'].value_counts(normalize=True) * 100
print('Porcentaje de colores en inglés:')
print(english_color_percentages)

plt.figure(figsize=(10, 7))
sns.barplot(x=english_color_percentages.values, y=english_color_percentages.index)
plt.title('Porcentaje de Colores en Inglés')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Color en Inglés')
plt.tight_layout()
plt.show()
```

```python
spanish_color_percentages = munsell['spanish_color'].value_counts(normalize=True) * 100
print('Porcentaje de colores en español:')
print(spanish_color_percentages)
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x=spanish_color_percentages.values, y=spanish_color_percentages.index)
plt.title('Porcentaje de Colores en Español')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Color en Español')
plt.tight_layout()
plt.show()
```

```python
tsimane_color_percentages = munsell['tsimane_color'].value_counts(normalize=True) * 100
print('Porcentaje de colores en Tsimane:')
print(tsimane_color_percentages)
```

```python
plt.figure(figsize=(10, 7))
sns.barplot(x=tsimane_color_percentages.values, y=tsimane_color_percentages.index)
plt.title('Porcentaje de Colores en Ingles')
plt.xlabel('Porcentaje (%)')
plt.ylabel('Color en tsimane')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

correlation_df = pd.DataFrame({
    'english': english_color_percentages,
    'spanish': spanish_color_percentages
}).fillna(0)

correlation_coefficient = correlation_df['english'].corr(correlation_df['spanish'])

plt.figure(figsize=(10, 7))
sns.scatterplot(x='english', y='spanish', data=correlation_df)
plt.title('Correlación de Porcentajes de Colores (Inglés vs. Español)')
plt.xlabel('Porcentaje de Colores en Inglés (%)')
plt.ylabel('Porcentaje de Colores en Español (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Coeficiente de correlación entre los porcentajes de colores (Inglés vs. Español): {correlation_coefficient:.2f}")
```

```python
correlation_df_en_ts = pd.DataFrame({
    'english': english_color_percentages,
    'tsimane': tsimane_color_percentages
}).fillna(0)

correlation_coefficient_en_ts = correlation_df_en_ts['english'].corr(correlation_df_en_ts['tsimane'])

plt.figure(figsize=(10, 7))
sns.scatterplot(x='english', y='tsimane', data=correlation_df_en_ts)
plt.title('Correlación de Porcentajes de Colores (Inglés vs. Tsimane)')
plt.xlabel('Porcentaje de Colores en Inglés (%)')
plt.ylabel('Porcentaje de Colores en Tsimane (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Coeficiente de correlación entre los porcentajes de colores (Inglés vs. Tsimane): {correlation_coefficient_en_ts:.2f}")
```

```python
correlation_df_es_ts = pd.DataFrame({
    'spanish': spanish_color_percentages,
    'tsimane': tsimane_color_percentages
}).fillna(0)

correlation_coefficient_es_ts = correlation_df_es_ts['spanish'].corr(correlation_df_es_ts['tsimane'])

plt.figure(figsize=(10, 7))
sns.scatterplot(x='spanish', y='tsimane', data=correlation_df_es_ts)
plt.title('Correlación de Porcentajes de Colores (Español vs. Tsimane)')
plt.xlabel('Porcentaje de Colores en Español (%)')
plt.ylabel('Porcentaje de Colores en Tsimane (%)')
plt.grid(True)
plt.tight_layout()
plt.show()

print(f"Coeficiente de correlación entre los porcentajes de colores (Español vs. Tsimane): {correlation_coefficient_es_ts:.2f}")
```
